```python
## Module 1 DAP
## Main Functions for import and review

## import pandas and numpy
import pandas as pd
import numpy as np


## load data
datasource = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

data = pd.read_csv(datasource,header = None, names = headers)

## df.columns = headers in case of later update
```
